# Determinant of Shifted Matrices

### Instructions:
1. Generate a 20x20 random matrix.
2. Impose a linear dependence
3. Shift the matrix by some lambda
4. Compute the absolute value of the determinant
5. Repeat 1000 times and take the average of step 4.
6. Plot determinant against lambda


```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate lambda values
lambdaValues = np.linspace(0,.1,30)
lambdaValues
```

```python
#empty zeroes array to house determinants
determinants = np.zeros(1000)
dets = np.zeros(len(lambdaValues))
```

```python
for det in range(len(lambdaValues)):
    for i in range(1000):
        m = np.random.randn(20,20)
        m[:,0] = m[:,1] #create linear dependence
        determinants[i] = abs( np.linalg.det( m + lambdaValues[det]*np.eye(20))) #shift the matrix by lambdaValues
    dets[det] = np.mean(determinants)
```

```python
plt.plot(lambdaValues, dets, 's-')
plt.xlabel('Fraction of identity for Shifting')
plt.ylabel('determinant')
plt.show
```
